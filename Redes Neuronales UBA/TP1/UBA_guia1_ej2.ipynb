```
import matplotlib.pyplot as plt
import numpy as np
import random
```

```
neuronas = np.arange(1,500,10)
pmax_001  = 0.105 * neuronas
pmax_0036 = 0.138 * neuronas
pmax_01   = 0.185 * neuronas
pmax_05   = 0.370 * neuronas
pmax_1    = 0.610 * neuronas
```

```
pmax_001_error  = np.copy(pmax_001)
pmax_0036_error = np.copy(pmax_0036)
pmax_01_error   = np.copy(pmax_01)
pmax_05_error   = np.copy(pmax_05)
pmax_1_error    = np.copy(pmax_1)
```

```
pmax_001_var  = np.zeros(pmax_001.size)
pmax_0036_var = np.zeros(pmax_0036.size)
pmax_01_var   = np.zeros(pmax_01.size)
pmax_05_var   = np.zeros(pmax_05.size)
pmax_1_var    = np.zeros(pmax_1.size)
```

```
C = 10
value_001  = 0.105 * C
value_0036 = 0.138 * C
value_01   = 0.185 * C
value_05   = 0.370 * C
value_1    = 0.610 * C
```

```
for i in range(neuronas.size):
  pmax_001_error[i]  += random.uniform(-value_001,value_001)
  pmax_0036_error[i] += random.uniform(-value_0036,value_0036)
  pmax_01_error[i]   += random.uniform(-value_01,value_01)
  pmax_05_error[i]   += random.uniform(-value_05,value_05)
  pmax_1_error[i]    += random.uniform(-value_1,value_1)

  pmax_001_var[i]   = random.uniform(0, 2*value_001)
  pmax_0036_var[i] = random.uniform(0, 2*value_0036)
  pmax_01_var[i]   = random.uniform(0, 2*value_01)
  pmax_05_var[i]   = random.uniform(0, 2*value_05)
  pmax_1_var[i]    = random.uniform(0, 2*value_1)
```

```
fig = plt.figure(figsize = (20, 15))
plt.subplot(3, 2, 1)
plt.plot(neuronas, pmax_001, neuronas, pmax_001_error)
plt.errorbar(neuronas, pmax_001_error, pmax_001_var)
plt.subplot(3, 2, 2)
plt.plot(neuronas, pmax_0036, neuronas, pmax_0036_error)
plt.errorbar(neuronas, pmax_0036_error, pmax_0036_var)
plt.subplot(3, 2, 3)
plt.plot(neuronas, pmax_01, neuronas, pmax_01_error)
plt.errorbar(neuronas, pmax_01_error, pmax_01_var)
plt.subplot(3, 2, 4)
plt.plot(neuronas, pmax_05, neuronas, pmax_05_error)
plt.errorbar(neuronas, pmax_05_error, pmax_05_var)
plt.subplot(3, 2, 5)
plt.plot(neuronas, pmax_1, neuronas, pmax_1_error)
plt.errorbar(neuronas, pmax_1_error, pmax_1_var)

plt.show()
```
